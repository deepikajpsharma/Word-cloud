```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("non_annex.csv", parse_dates=["Date"])
```

When you saved the dataframe to CSV, the lists got saved as a string, so we can read using literal_eval function

```python
from ast import literal_eval
df['Processed'] = df['Processed'].apply(literal_eval)
```

```python
stopwords = ['COP26']
words = ''
for item in df['Processed']:
    words += " ".join(list(item))
```

```python
words
```

```python
wordcloud = WordCloud(width = 1024, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(words)
 
# plot the WordCloud image                      
plt.figure(figsize = (10, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
plt.savefig("wordcloud.jpg")
plt.show()
```
